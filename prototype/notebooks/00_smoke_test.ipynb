# SSC Prototype — Smoke Test (Phase 1 → 2)

Loads the ontology `ssc.ttl`, materializes OWL-RL inferences, and lists the vocabulary.
Run the cells top to bottom. Requires `rdflib` and `owlrl` (see `../requirements.txt`).

```python
from pathlib import Path
from rdflib import Graph

ONTO = Path('../ontology-ssc/ssc.ttl')
g = Graph()
g.parse(ONTO, format='turtle')
print(f'Loaded {len(g)} triples from {ONTO}')
```

```python
import owlrl
before = len(g)
owlrl.DeductiveClosure(owlrl.OWLRL_Semantics).expand(g)
print(f'Triples: {before} -> {len(g)} after OWL-RL inference')
```

```python
OWL = 'http://www.w3.org/2002/07/owl#'
classes = sorted(str(r[0]).split('#')[-1] for r in g.query(
    'SELECT ?c WHERE { ?c a <%sClass> }' % OWL))
props = sorted(str(r[0]).split('#')[-1] for r in g.query(
    'SELECT ?p WHERE { ?p a <%sObjectProperty> }' % OWL))
print(f'{len(classes)} classes:', classes)
print()
print(f'{len(props)} object properties:', props)
```

# Phase 4 — Reusable query library

Uses the helper `src/ssc_graph.py` to load ontology + data, run OWL-RL inference, and execute every saved competency-question query in `../queries/`. This confirms the reasoning end-to-end on the seed data.

```python
import sys; sys.path.append('../src')
import ssc_graph as kg
from pathlib import Path

g = kg.load(['../data/seed.ttl'])   # ontology + seed, with inference
print(len(g), 'triples (after inference)')
```

```python
for f in sorted(Path('../queries').glob('*.rq')):
    print(f'\n== {f.stem} ==')
    rows = kg.run_file(g, f)
    if not rows:
        print('  (no results)')
    for row in rows:
        print('  ', ' | '.join(kg.short(x) for x in row))
```

## What you have now
A saved, reusable set of competency-question queries plus a one-line loader. Swap `../data/seed.ttl` for real build/intelligence facts later (Phase 5+) and the same queries work unchanged.
